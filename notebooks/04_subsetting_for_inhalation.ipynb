```python
def filter_data(df):
    
    mask = (df['study_length'].isin(['subacute','subchronic', 'repeat dose','chronic', 'reproductive developmental', 'reproductive', 'developmental', 'multigenerational reproductive'])) & (df['exposure_route'] == 'inhalation') & (df['toxval_type'].isin(['NOAEL', 'NOEL', 'NOAEC', 'NOEC'])) & (df['species'] == 'rodent') 
    
    return df[mask]
```

```python
def representative(s):
    if len(s) ==1:
        return float(s)
    while len(s) >1:
        q75 = pd.Series(s.quantile(0.75))
        q25 = pd.Series(s.quantile(0.25))
        iqr = q75-q25
        pt_lower = q25 - (iqr*1.5)
        pt_higher = q75 + (iqr*1.5)
         
        newset = pd.Series([e for e in s if not (e < pt_lower).any() if not (e > pt_higher).any()])
        
        #q25out = newset.quantile(0.25)
        
        #front = pd.Series([e for e in newset if e< q25out])
        
        return newset.min()
```
